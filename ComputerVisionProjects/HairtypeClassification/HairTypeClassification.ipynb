```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale = 1/255, 
                             rotation_range=10, 
                             width_shift_range=0.1, 
                             shear_range=0.1,
                             zoom_range=0.1, 
                             horizontal_flip = True,
                             fill_mode='nearest', 
                             validation_split=0.2
                            )
```

```python
train_gen = datagen.flow_from_directory(img_dir, 
                                        target_size = (128, 128), 
                                        class_mode='binary', 
                                        batch_size = 20,
                                        subset = "training")

val_gen = datagen.flow_from_directory(img_dir, 
                                        target_size = (128, 128), 
                                        class_mode='binary', 
                                        batch_size = 20,
                                        subset = "validation")
```

```python
train_dim = next(train_gen)
print("Dimensions of a training sample",train_dim[0].shape)

# checking the new dimension of the validation data, which should include the batch size of 16 and new pixel size of 32 instead of 224 
val_dim = next(val_gen)
print("Dimensions of a validation sample",val_dim[0].shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, BatchNormalization, Flatten, Lambda, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.optimizers import RMSprop, Adam
```

```python
model = Sequential([
    Lambda(lambda x: (x-tf.math.reduce_mean(x))/tf.math.reduce_std(x), input_shape=(128, 128, 3)),
    Conv2D(32, (5,5), activation='relu'),
    BatchNormalization(),
    Conv2D(32, (4,4), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(32, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPool2D(2, 2),
    
    Conv2D(64, (5,5), activation='relu'),
    BatchNormalization(),
    Conv2D(64, (4,4), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPool2D(2, 2),
    
    Conv2D(128, (5,5), activation='relu'),
    BatchNormalization(),
    Conv2D(128, (4,4), activation='relu'),
    BatchNormalization(),
    Conv2D(128, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPool2D(2, 2),
    
    Flatten(), 
    Dense(512, activation = "relu", kernel_regularizer=l2(0.005)),
    Dense(256, activation = "relu", kernel_regularizer=l2(0.005)),
    Dense(128, activation = "relu", kernel_regularizer=l2(0.005)),
    Dense(1, activation = "sigmoid") 
]) 

model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
model.summary()
```

```python
lr_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience=2, factor=0.5, min_lr=0.00001)
```

```python
history = model.fit(
            train_gen,
            epochs=50,
            validation_data=val_gen,
            verbose=2,
            callbacks = lr_reduction
            )
```
